**IMAGE - 1:**



```python
from IPython.core.display import display_jpeg
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files


# Put your code here:
i = cv2.imread("/content/Image_1.jpg",cv2.IMREAD_COLOR)
cv2_imshow(i)
(rows,cols) = i.shape[:2]
i1 = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
res = cv2.warpAffine(i,i1,(cols,rows))
i2 = cv2.cvtColor(res, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(i2,(25,36,25),(86,255,255))
imask = mask>0
green = np.zeros_like(i,np.uint8)
green[imask] = i[imask]
cv2_imshow(green)
```

**IMAGE - 1:**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Put your code here:
i = cv2.imread("/content/1.jpeg",cv2.IMREAD_COLOR)
cv2_imshow(i)
[thresh,im_bw] = cv2.threshold(i,0,200,cv2.THRESH_BINARY,cv2.THRESH_OTSU)
i2 = cv2.cvtColor(im_bw, cv2.COLOR_BGR2GRAY)
kernel = np.ones((3,3), np.uint8)
i3 = cv2.erode(i2, kernel, iterations=2)
cv2_imshow(i3)

```
